(5) (Random walk with drift) Use a biased coin to simulate a random
walk of 30 steps on the line. If the coin falls heads (H), take one step to
the right, if it lands tails (T ), take one step left. After 30 steps, note the
final position. Take Pr(H)=0.6 and Pr(T )=0.4. (a) Plot a sample path.
(b) Make a histogram for 200 such random walks. (c) Report the sample
mean. (d) Report the sample variance. (What should these be exactly?)

```python
import numpy as np
import matplotlib.pyplot as plt
```

For this exercise, I translated (and slightly modified) the Matlab code from this source into Python:
    https://www.coursehero.com/file/8649156/HWK2Solutions/

```python
# Part A

z = 2*(np.random.rand(1,30)<0.6)-1 # increments; 1s and -1s
z=list(z) 

x=np.zeros(31)
x=list(x)

for i in range(31):
    x[i]=x[i-1]+z[0][i-1] #adding the increments                                  

plt.plot(np.linspace(1,31,31),x)
plt.xlabel('Number of Steps')
plt.ylabel('Position')
plt.title('Sample Path')
plt.show()
```

```python
# Part B

ending_positions = []

for n in range(200):
    z = 2*(np.random.rand(1,30)<0.6)-1 # a vector of increments (entries +1,-1)
    z=list(z) 

    x=np.zeros(31)
    x=list(x)

    for i in range(31):
        x[i]=x[i-1]+z[0][i-1] #adding the increments to obtain a sample path  
        
    ending_positions.append(x[-1])

plt.hist(ending_positions)
plt.title('Histogram of Ending Positions')
plt.show()
print('Sample mean: ', np.mean(ending_positions))
print('Sample variance: ', np.var(ending_positions))
```

```python
# Translated textbook Matlab code to Python
import random

R=[]
R.append(100)

i = 0

while R[i]>0 and R[i]<2100:
    i=i+1
    w = random.randint(0,1) # random value of 0 or 1, gambler wins
    w = 2*w - 1 #random value of +/-1
    R.append(R[i-1]+ w) # gamblers new fortune

# print(R)
x = range(len(R))
plt.plot(x,R) # plot R against its index
plt.show()
```

```python
'''
Run for investigating running time
'''

T=[]

for j in range(20):
    R=[]
    R.append(100)
    i = 0
    while R[i]>0 and R[i]<2100:
        i=i+1
        w = random.randint(0,1) # random value of 0 or 1, gambler wins
        w = 2*w - 1 #random value of +/-1
        R.append(R[i-1]+ w) # gamblers new fortune

    T.append(i)
    
plt.hist(T)
plt.show()
```
