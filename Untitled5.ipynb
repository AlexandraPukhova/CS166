```python
import matplotlib
matplotlib.use('TkAgg')

import pylab as PL
import numpy as np

import random as RD
import scipy as SP

RD.seed()

width = 50
height = 50
initProb = 0.01
infectionRate = 0.9
regrowthRate = 0.15

def init():
    global time, config, nextConfig, density_lst

    time = 0
    
    density_lst = [] # added density list for healthy hosts
    
    config = SP.zeros([height, width])
    for x in range(width):
        for y in range(height):
            if RD.random() < initProb:
                state = 2
            else:
                state = 1
            config[y, x] = state

    nextConfig = SP.zeros([height, width])

def draw():
    
    infected = np.count_nonzero(config==2)
    healthy = np.count_nonzero(config==1)
    density_lst.append(healthy/infected)
    
    PL.cla()
    
    PL.subplot(1,2,1)
    PL.pcolor(config, vmin = 0, vmax = 2, cmap = PL.cm.jet)
    PL.axis('image')
    PL.title('t = ' + str(time))
    
    PL.subplot(1,2,2)
    PL.plot(range(len(density_lst)), density_lst) # adding the time vs density plot

def step():
    global time, config, nextConfig

    time += 1

    for x in range(width):
        for y in range(height):
            state = config[y, x]
            if state == 0:
                for dx in range(-1, 2):
                    for dy in range(-1, 2):
                        if config[(y+dy)%height, (x+dx)%width] == 1:
                            if RD.random() < regrowthRate:
                                state = 1
            elif state == 1:
                for dx in range(-1, 2):
                    for dy in range(-1, 2):
                        if config[(y+dy)%height, (x+dx)%width] == 2:
                            if RD.random() < infectionRate:
                                state = 2
            else:
                state = 0

            nextConfig[y, x] = state

    config, nextConfig = nextConfig, config
    
    
import pycxsimulator
pycxsimulator.GUI().start(func=[init,draw,step])
```

* What happens if the infection probability is changed?
If the infection probability is changed(e.g. to 10%), the population of infected hosts dies out and the state converges to only having healthy hosts. The opposite works for when we increase the proportion of infected hosts above 0.85, e.g. to 0.9 

* In what conditions will both hosts and pathogens co-exist? In what conditions can hosts exterminate pathogens? 
Both hosts and pathogens co-exist at the infection rate of 85%.
