```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

def q(k):
    from scipy.optimize import root
    return root(lambda q: q - np.exp(k * (q - 1)), 0).x[0]

n = 1000
k= np.linspace(1, 10, 100)
q = np.array([q(_) for _ in k])
llc_size= (1-q)*n

plt.plot(k, llc_size)
plt.xlabel('Average degree k')
plt.ylabel('LLC size')
plt.title('LLC size vs Average degree k')
plt.show()
```
